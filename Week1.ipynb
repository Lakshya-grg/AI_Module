4.2 Experiment: How Fast is Numpy? In this exercise, you will compare the performance and implementation of operations using plain Python lists (arrays) and NumPy arrays. Follow the instructions:

Element-wise Addition: • Using Python Lists, perform element-wise addition of two lists of size 1, 000, 000. Measure and Print the time taken for this operation.


```python
import time
import random

size = 1_000_000
list1 = [random.random() for _ in range(size)]
list2 = [random.random() for _ in range(size)]

start_time = time.time()

result = [list1[i] + list2[i] for i in range(size)]

end_time = time.time()

print("Time taken (Python lists):", end_time - start_time, "seconds")
```

 Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this operation.

```python
import numpy as np
import time

size = 1_000_000
arr1 = np.random.rand(size)
arr2 = np.random.rand(size)

start_time = time.time()

result = arr1 + arr2

end_time = time.time()

print("Time taken (NumPy arrays):", end_time - start_time, "seconds")
```

Matrix Multiplication • Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.

```python
n = 1000

A = [[random.random() for _ in range(n)] for _ in range(n)]
B = [[random.random() for _ in range(n)] for _ in range(n)]

C = [[0 for _ in range(n)] for _ in range(n)]

start_time = time.time()

for i in range(n):
    for j in range(n):
        total = 0
        for k in range(n):
            total += A[i][k] * B[k][j]
        C[i][j] = total

end_time = time.time()

print("Time taken (Python lists - 1000x1000 matrix multiplication):", end_time - start_time, "seconds")
```

Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000. Mea- sure and print the time taken for this operation.

```python
import numpy as np
import time

size = 1000   # define size first!

A_np = np.ones((size, size))
B_np = np.ones((size, size))

start = time.time()
C_np = np.dot(A_np, B_np)   # or A_np @ B_np
end = time.time()

print("NumPy Matrix Multiplication Time:", end - start, "seconds")
```
